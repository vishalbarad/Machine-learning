```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
```

```python
from sklearn.cluster import DBSCAN
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
x,label = make_moons(n_samples=200,noise=0.16)#here we add noise to see difference
x[:5]
```

```python
label
```

```python
model = DBSCAN(eps=0.25,min_samples=10)
model.fit(x)
```

```python
model.labels_ #here -1 is indicates error or noise
```

```python
plt.scatter(x[:,0],x[:,1],c=model.labels_)#here we added a noise so that we get this type of model with outlier(noise)
```

```python
plt.scatter(x[:,0],x[:,1],c=label)#this is a original dataset label model
```
